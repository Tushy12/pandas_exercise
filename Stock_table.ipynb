```python
import numpy as np
import pandas as pd
import os

# package to extract data from various Internet sources into a DataFrame
# make sure you have it installed
import pandas_datareader.data as web

import datetime as dt
import seaborn as sns
```

```python
df = web.DataReader("AAPL", "av-daily", start='1/1/2015',
                   end='1/1/2021',api_key='key')
```

```python
df['stock'] = 'AAPL'
```

```python
df
```

```python
df1 = web.DataReader("TSLA", "av-daily", start='1/1/2015',
                   end='1/1/2021',api_key='key')
df1['stock'] = 'TSLA'
df1
```

```python
df2 = web.DataReader("IBM", "av-daily", start='1/1/2015',
                   end='1/1/2021',api_key='key')
df2['stock'] = 'IBM'
df2
```

```python
df3 = web.DataReader("MSFT", "av-daily", start='1/1/2015',
                   end='1/1/2021',api_key='key')
df3['stock'] = 'MSFT'
df3
```

```python
all_data = pd.concat([df,df1,df2,df3], ignore_index=False)
```

```python
all_data
```

```python

all_data.set_index('stock',append=True, inplace=True)
all_data
```

```python
vol = all_data.drop(columns=['open','high','low','close'])
vol
```

```python
all_data_plot = pd.concat([df,df1,df2,df3], ignore_index=False)
sns.lineplot(data=all_data_plot,x=all_data_plot.index, y='close', palette='husl', hue='stock',
             )
```

```python
all_data_plot['index'] = range(1,6045)
```

```python
year = all_data_plot.index
```

```python
all_data_plot.set_index('index', inplace=True)
all_data_plot['Year'] = year
all_data_plot
```

```python
vol = all_data_plot.drop(columns=['open','high','low','close'])
vol['Year'] = pd.to_datetime(vol['Year'])
vol['Week'] = vol['Year'].dt.isocalendar().week
vol.head(50)
```

```python
table = pd.pivot_table(vol, values='volume', index=['Year','Week'],
                       columns=['stock'], aggfunc=np.sum)
```
