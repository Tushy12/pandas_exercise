#### Question 1: Legal_drinkers?

```python
def legal_drink(student):
    if student > 17:
        return 1
    else:
        return 0

all_data['legal_drinker'] = all_data['age'].apply(legal_drink)
all_data.head()
```

```python
csv_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv'
df = pd.read_csv(csv_url)
df.head()
```

```python
np.dtype(df['Year'])
```

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df.head()
```

```python
df = df.set_index('Year', drop = True)

df.head()
```

```python
df = df.resample('10AS').sum()
```
